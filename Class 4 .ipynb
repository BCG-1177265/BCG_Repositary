```python
import seaborn as sns
```

```python
%ls datasets/
```

```python
df = pd.read_csv('datasets/Cartwheeldata.csv')
df.head()
```

```python
sns.lmplot(x = 'Wingspan',
           y = 'CWDistance',
           data = df,
           fit_reg = False,
           hue = 'Gender'
          )

plt.title("Scatter Plot of CW Distance vs. Wingspan", fontsize = 14)
plt.show()
```

```python
sns.swarmplot(x = 'Gender', 
             y = 'CWDistance',
             data = df
            )
plt.show()
```

```python
sns.boxplot(data = df.loc[:,['Age', 'Height', 'Wingspan', 'CWDistance', 'Score']])

plt.xticks(rotation = 45, fontsize = 12)
plt.yticks(fontsize = 12)

plt.title('Box plot reprensetantion of the entire data', fontsize = 15)
plt.show()
```

### Box plot of Male data

```python
sns.boxplot(data = df.loc[df['Gender']== 'M' ,['Age', 'Height', 'Wingspan', 'CWDistance', 'Score']])

plt.xticks(rotation = 45, fontsize = 12)
plt.yticks(fontsize = 12)

plt.title('Box plot reprensetantion of the Male data', fontsize = 15)
plt.show()
```

### Graphs side by side (scores by gender)

```python
fig = plt.figure()

ax1 = fig.add_subplot(3,2,1) # (rows, columns, position)
sns.boxplot(data = df.loc[df['Gender'] == 'M',['Score']])


ax2 = fig.add_subplot(3,2,2) 
sns.boxplot(data = df.loc[df['Gender'] == 'F',['Score']])


ax3 = fig.add_subplot(3,2,5)
sns.boxplot(data = df.loc[df['Gender'] == 'M',['Score']])


ax4 = fig.add_subplot(3,2,6)
sns.boxplot(data = df.loc[df['Gender'] == 'F',['Score']])


ax1.title.set_text('Male box plot')
ax2.title.set_text('Female box plot')
ax3.title.set_text('Male box plot')
ax4.title.set_text('Female box plot')
plt.show()
```
