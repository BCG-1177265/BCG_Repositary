# Spotify Top 50 Songs of 2020 Dataset
## This data set is about Spotify Music
## This data is provided by Kaggle
## This data set contains the top 50 tracks listened on spotify in 2020


```python
df = pd.read_csv('newdata/spotifytoptracks.csv')
df
```

### Eliminar variables innecesarias para mi estudio.

```python
df.drop(['Unnamed: 0', 'track_id'],
        axis = 1, 
        inplace = True
       )
```

## Variables de interes:
* 'artist' --> El artista de la cancion, es una variable categorica.
* 'album' --> El album que pertence la cancion, es una variable categorica.
* 'track_name' --> El nombre de la cancion, es una variable categorica.
* 'energy' --> La energia de la cancion, es una variable numerica.
* 'danceability' --> Que tan bailable es la cancion, es una variable numerica.
* 'acousticness'--> La acustica de cada cancion, es una variable numerica.
* 'instrumentalness'--> Instrumentalidad de la cancion, es una variable numerica.
* 'liveness' --> Que tan viva es la cancion, es una variable numerica.
* 'tempo' --> El tempo de la cancion, es una variable numerica.
* 'duration_ms' --> La duracion en milisegundos de cada cancion, es una variable numerica.
* 'genre' --> El genero que pertenece cada cancion, es una variable categorica.

## Danceability of each song by artist
* Usando una tabla pivot, demuestra que tan bailable es cada cancion de cada artista, se puede ver si el artista tiene mas de una cancion

```python
pd.pivot_table(df, 
               index = ['artist', 'track_name'],
               values = ['danceability'],
               margins = True,
              )
```

* En conclusion las canciones tienen una baileabilidad de 0.71672

## Average energy of the songs 

```python
sns.swarmplot(x = 'track_name', 
             y = 'energy',
             data = df
            )
plt.xticks(rotation = 90, fontsize = 10)
plt.yticks(rotation = 45, fontsize = 10)

plt.show()
```

## How many songs of the same artist were in the top 50?

```python
sns.countplot(x = 'artist', data = df)

plt.xticks(rotation = 90, fontsize = 10)


plt.show()
```

## "Scatter Plot of Duration_ms vs. Danceability by Genre"

```python
sns.lmplot(x = 'duration_ms',
           y = 'danceability',
           data = df,
           fit_reg = False,
           hue = 'genre'
          )
plt.title("Scatter Plot of Duration_ms vs. Danceability by Genre", fontsize = 14)
plt.xticks(rotation = 45, fontsize = 10)
plt.show()
```
