## Variables de interes:
* 'artist' --> El artista de la cancion, es una variable categorica.
* 'album' --> El album que pertence la cancion, es una variable categorica.
* 'track_name' --> El nombre de la cancion, es una variable categorica.
* 'energy' --> La energia de la cancion, es una variable numerica.
* 'danceability' --> Que tan bailable es la cancion, es una variable numerica.
* 'acousticness'--> La acustica de cada cancion, es una variable numerica.
* 'instrumentalness'--> Instrumentalidad de la cancion, es una variable numerica.
* 'liveness' --> Que tan viva es la cancion, es una variable numerica.
* 'tempo' --> El tempo de la cancion, es una variable numerica.
* 'duration_ms' --> La duracion en milisegundos de cada cancion, es una variable numerica.
* 'genre' --> El genero que pertenece cada cancion, es una variable categorica.

## Average energy of the songs 
* Se busca ver visualmente la energia de cada cancion, por lo que usaremos una gradica para mostrar los datos

```python
sns.swarmplot(x = 'track_name', 
             y = 'energy',
             data = df
            )
sns.set(rc={'figure.figsize':(13,10)})
plt.xticks(rotation = 90, fontsize = 13)
plt.yticks(rotation = 45, fontsize = 13)

plt.show()
```

#### Conclusi칩n --> Ahora se puede ver la dispersion de datos de como cambia la energia de cada cancion.

## How many songs of the same artist were in the top 50?

```python
sns.countplot(x = 'artist', data = df)
sns.set(rc={"figure.figsize":(13,10)})

plt.xticks(rotation = 90, fontsize = 10)


plt.show()
```

#### Conclusion --> se puede ver que 'Dua Lipa', 'Billie Eilish' y 'Travis Scott' tuvieron 3 de sus canciones en la lista de top 50, mientras que 'Harry Styles'. 'Lewis Capaldi', 'Post Malone' y 'Justin Bieber' tuvieron 2 de sus canciones en la lista.

## "Scatter Plot of Duration_ms vs. Danceability by Genre"
* Se busca conocer si la duracion de una cancion tiene un impacto en la manera que se baila, ademas esta categorizado por el genero de la cancion.

```python
sns.lmplot(x = 'duration_ms',
           y = 'danceability',
           data = df,
           fit_reg = False,
           hue = 'genre'
          )
sns.set(rc={'figure.figsize':(13,10)})

plt.title("Scatter Plot of Duration_ms vs. Danceability by Genre", fontsize = 14)
plt.xticks(rotation = 45, fontsize = 10)
plt.show()
```

#### Conclusion --> Por lo visto la mayoria de las canciones tienen una duracion corta, en comparacion con los ultimos dos puntitos que muestra que la cancion que mas dura tiene una baileabilidad alta y es del genero 'Dance/Electronic'

## Difference between top son and a random song in the list
* Se va a visualizar si la cancion mas escuchada tiene alguna diferencia en cuando a energy, danceability, key, loudness y accousticness

```python
nf = df.loc[['Blinding Lights','lovely (with Khalid)'], 'energy':'acousticness']

nic = nf.transpose()

nic.plot(figsize = (15,10))

plt.xlabel('Songs', fontsize =17)
plt.xticks(fontsize = 10)

plt.ylabel('Energy', fontsize = 17)
plt.yticks(fontsize = 14)

plt.title('Comparison top song (Blinding Lights) vs Lovely (with Khalid)', fontsize=17)

plt.show()
```

#### Conclusion --> Por lo visto la cancion 'Blinding Lights' tiene una mayor 'key', menor 'loudness' y mayor 'acousticness', mientras que las otras variables permanecen similar.
