```python
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

df1 = pd.DataFrame(df['sistol'])
df2 = pd.DataFrame(df['diastol'])

fig, axes = plt.subplots(1, 2)

df1.hist( ax=axes[0])
df2.hist( ax=axes[1])

plt.show()
```
