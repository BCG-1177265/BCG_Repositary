# Description of the datasets



* This datasets represents how many people from a country reached Canada over the years

### My variables of interest
* 'Country' describes the origin of the people, this is a Categorical vairable.
* 'Continent' represents the continent of the origin of the people, this is a Categorical vairable.
* 'Region' is the region of the ...
* This data set includes the years from '1908 - 2013'

```python
df.columns = list(map(str, df.columns))
df.columns
```

```python
df.set_index('Country', inplace = True)
df.head()
```

```python
df['Total'] = df.iloc [:,
                       2:].sum(axis=1)
df.head(2)
```

```python
j_data = df.loc['Japan', '1980':'2013']
j_data.plot()
```

```python
ic_data = df.loc[['India','China'], '1980':'2013']

nic = ic_data.transpose()

nic.plot(figsize = (8,6))

plt.xlabel('Years', fontsize =17)
plt.xticks(fontsize = 14, rotation = 45)

plt.ylabel('Number of immigrants', fontsize = 17)
plt.yticks(fontsize = 14)

plt.title('Immigration to Canada from India and China during 1980 to 2013', fontsize=17)

plt.show()
```

## Conclusion

We see a similar pattern from China & India.
